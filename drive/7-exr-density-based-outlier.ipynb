Exercise 1:

Abstract: Apply the DBSCAN density-based clustering algorithm on the Iris dataset to analyze the distribution of data points. Identify dense regions that form clusters and detect possible outliers in the dataset based on the parameters epsilon (ε) and minimum points (MinPts). Visualize the resulting clusters and noise points using appropriate Python libraries.

Dataset:
UCI Repository link: https://archive.ics.uci.edu/ml/datasets/iris

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
file_path = "/content/iris.data"
```

```python
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
df = pd.read_csv(file_path, header=None, names=columns)
```

```python
X = df.iloc[:, :-1].values
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
db = DBSCAN(eps=0.8, min_samples=5)
clusters = db.fit_predict(X_scaled)
```

```python
print("Clusters:", len(set(clusters)) - (1 if -1 in clusters else 0))
print("Outliers:", list(clusters).count(-1))
```

```python
plt.scatter(X_scaled[:, 2], X_scaled[:, 3], c=clusters, cmap='viridis')
plt.title("DBSCAN on Iris Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
```

```python
plt.scatter(X_scaled[clusters == -1, 0],
            X_scaled[clusters == -1, 1],
            color='red', marker='x', s=100)

plt.show()
```

Exercise 2:

Abstract: Use the DBSCAN algorithm to perform density-based clustering on the Breast Cancer Wisconsin dataset. Analyze the dataset to identify clusters of similar data points and detect observations that behave as outliers or noise. Visualize the clusters and detected outliers to understand the density distribution of the data.

Dataset: https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
file_path = "/content/wdbc.data"
```

```python
columns = ['id', 'diagnosis'] + [f'feature_{i}' for i in range(30)]
df = pd.read_csv(file_path, header=None, names=columns)
```

```python
X = df.iloc[:, 2:].values
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
db = DBSCAN(eps=4, min_samples=5)
clusters = db.fit_predict(X_scaled)
```

```python
print("Clusters:", len(set(clusters)) - (1 if -1 in clusters else 0))
print("Outliers:", list(clusters).count(-1))
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='viridis')
plt.title("DBSCAN on Breast Cancer Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
```

```python
plt.scatter(X_scaled[clusters == -1, 0],
            X_scaled[clusters == -1, 1],
            color='red', marker='x', s=100)

plt.show()
```
