Exercise 1:\
Abstract: In this exercise, a paired sample t-test is conducted to evaluate the impact of backend optimization techniques on website response time. Performance measurements collected before and after optimization for the same test cases are analyzed to determine whether the observed reduction in response
time is statistically significant. This exercise helps students develop analytical skills in performance evaluation and evidence-based system optimization using inferential statistics.

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
data = {
    "Test_Case": ["T1","T2","T3","T4","T5","T6","T7","T8"],
    "Response_Time_Before": [420, 450, 410, 470, 440, 460, 430, 445],
    "Response_Time_After": [360, 390, 355, 400, 380, 395, 370, 385]
}
df = pd.DataFrame(data)
df
```

```python
df["Difference"] = df["Response_Time_Before"] - df["Response_Time_After"]
df
```

```python
t_statistic, p_value = stats.ttest_rel(
    df["Response_Time_Before"],
    df["Response_Time_After"]
)
t_statistic, p_value
```

```python
print("Paired Sample t-test Results")
print("----------------------------")
print(f"t-statistic: {t_statistic:.4f}")
print(f"p-value: {p_value:.4f}")
```

```python
alpha = 0.05
if p_value < alpha:
    print("Conclusion: The reduction in response time is statistically significant.")
else:
    print("Conclusion: The reduction in response time is NOT statistically significant.")
```

Null Hypothesis (H₀): There is no significant difference in the website response time before and after applying backend optimization techniques. Any
observed reduction in response time is due to random variation.\
Alternative Hypothesis (H₁): There is a significant reduction in the website response time after applying backend optimization techniques.\
The paired sample t-test compares response times measured before and after backend optimization for the same test cases. Since the
calculated p-value is less than the significance level (α = 0.05), the null hypothesis is rejected. This indicates that backend optimization
techniques have significantly reduced website response time, proving their effectiveness through statistical evidence.

Exercise 2:\
Abstract: This exercise involves applying a one-sample t-test to analyze daily electricity consumption recorded from smart energy meters. The objective is to examine whether the average energy usage of households significantly differs from a predefined standard consumption level. Students will analyze the data by framing appropriate hypotheses, calculating the test statistic, and interpreting the results to support datadriven decisions related to energy efficiency and consumption monitoring.

```python
data = {
    "Household_ID": ["H1","H2","H3","H4","H5","H6","H7","H8","H9","H10"],
    "Daily_Energy_Consumption": [11.2, 10.8, 11.5, 10.6, 11.0, 11.3, 10.9, 11.1, 10.7, 11.4]
}
df = pd.DataFrame(data)
df
```

```python
population_mean = 11  # kWh
t_statistic, p_value = stats.ttest_1samp(
    df["Daily_Energy_Consumption"],
    population_mean
)
t_statistic, p_value
```

```python
print("One-Sample t-test Results")
print("-------------------------")
print(f"t-statistic: {t_statistic:.4f}")
print(f"p-value: {p_value:.4f}")
```

```python
alpha = 0.05
if p_value < alpha:
    print("Conclusion: The average daily energy consumption significantly differs from the assumed mean.")
else:
    print("Conclusion: The average daily energy consumption does NOT significantly differ from the assumed mean.")
```

Null Hypothesis (H₀): The average daily energy consumption of households recorded by smart energy meters is equal to the assumed standard consumption
level of 11 kWh.\
Alternative Hypothesis (H₁): The average daily energy consumption of households is different from the assumed standard consumption level of 11 kWh.\
A one-sample t-test was performed to compare the mean daily energy consumption of households with the assumed population mean
of 11 kWh. The calculated p-value is greater than the significance level (α = 0.05), hence the null hypothesis is not rejected. This
indicates that the average daily electricity consumption does not significantly differ from the standard consumption level. Therefore,
household energy usage is consistent with the expected benchmark, supporting effective energy monitoring.
