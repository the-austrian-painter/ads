```python
Pre_Post=pd.read_excel('/content/Pre_Post_Score.xlsx')
Pre_Post.head()
```

```python
Two_sample_result=stats.ttest_rel(Pre_Post['Pre_Score'],Pre_Post['Post_Score'])
Two_sample_result
```

H0 (Null Hypothesis): there is no significance difference between prescore and post score \
H1 (Alternative Hyposthesis) :there is significance difference between prescore and post score \
Hence using two sample Ttest we have calculated that T(tab) is greater than T(cal) H0 is accepted
