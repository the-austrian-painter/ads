```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_excel('/content/Party_Data.xlsx')
```

```python
df
```

```python
df['Time'].describe()
```

Measures of Central Tendency

```python
m=df['Time'].mean()
print(m)
```

```python
med=df['Time'].median()
print(med)
```

```python
mod=df['Time'].mode()
print(mod)
```

```python
harmean=stats.hmean(df['Time'])
print(harmean)
```

Measures of Dispersion

```python
m1=df['Time'].max()
m2=df['Time'].min()
range=m1-m2
print(range)
```

```python
Q3=df['Time'].quantile(0.75)
Q1=df['Time'].quantile(0.25)
IQR=Q3-Q1
print(IQR)
```

```python
df.boxplot(['Time'])
```

```python
s=df['Time'].std()
print(s)
```

```python
s1=df['Time'].var()
print(s1)
```

```python
cv=s/m
print(cv)
```

```python
dq=harmean/m
print(dq)
```

Measures of Shape

```python
skew=df['Time'].skew()
print(skew)
```

```python
kurtosis=df["Time"].kurtosis()
print(kurtosis)
```

```python
zscore=stats.zscore(df['Time'])
print(zscore)
```
