Exercise 1: Explore the descriptive statistics on the given dataset.\
Dataset: In this experiment, a fictitious data of Body Mass Index (BMI) containing 10 observations and 5 variables is used. The dataset contains Height, Weight, Age, BMI, and Gender columns.\
Abstract:

*   What is the typical BMI level of the study group, and how much variability exists in body mass across individuals falling within the clinical risk ranges of 25–27 (pre-diabetic), 28–30 (hypertensive–cholesterol), and ≥30 (obesity–cardiac)?

*   Does the distribution of BMI exhibit asymmetry and heavy-tailed behavior due to the presence of individuals in the extreme obesity category (BMI ≥ 30), indicating concentrated cardiac risk within the group?

*   Are any individuals statistically abnormal based on BMI Z-scores, particularly in the obesity range (BMI ≥ 30), and do these individuals emerge as true medical risk outliers capable of distorting overall group health assessment?

*   Which individual BMI values act as dominant health-risk contributors—analogous to high-cost late arrivals in event planning—by disproportionately inflating the group’s average BMI and
overall disease-risk perception?

*   If preventive health planning must be based on a single robust statistical measure, should the group intervention strategy rely on mean BMI or median BMI to minimize misclassification caused by extreme obesity values (BMI ≥ 30)?

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv('/content/bmi.csv')
df.columns = df.columns.str.strip().str.upper()
df.head()
```

```python
print("Columns in dataset:")
print(df.columns)
print("\nDataset info:")
df.info()
```

```python
mean_bmi = df["BMI"].mean()
median_bmi = df["BMI"].median()
print("Mean BMI:", mean_bmi)
print("Median BMI:", median_bmi)
```

```python
std_bmi = df["BMI"].std()
cv_bmi = (std_bmi / mean_bmi) * 100
print("Standard Deviation of BMI:", std_bmi)
print("Coefficient of Variation (%):", cv_bmi)
```

```python
pre_diabetic = df[(df["BMI"] >= 25) & (df["BMI"] <= 27)]
hypertensive = df[(df["BMI"] > 27) & (df["BMI"] <= 30)]
obesity = df[df["BMI"] >= 30]
print("Pre-diabetic (25–27) count:", len(pre_diabetic))
print("Hypertensive (28–30) count:", len(hypertensive))
print("Obesity (≥30) count:", len(obesity))
```

```python
skewness = stats.skew(df["BMI"])
kurtosis = stats.kurtosis(df["BMI"])
print("Skewness of BMI:", skewness)
print("Kurtosis of BMI:", kurtosis)
```

```python
df["BMI_ZSCORE"] = stats.zscore(df["BMI"])
df[["BMI", "BMI_ZSCORE"]]
```

```python
outliers = df[np.abs(df["BMI_ZSCORE"]) > 2]
print("Statistical Outliers based on BMI Z-score:")
outliers
```

```python
high_risk = df[df["BMI"] >= 30][["BMI", "BMI_ZSCORE"]]
print("High-risk obesity contributors:")
high_risk
```

```python
print("\nDecision for preventive health planning:")
if abs(mean_bmi - median_bmi) > 0.5:
    print("Median BMI is preferred due to influence of extreme obesity values.")
else:
    print("Mean BMI is acceptable as data is fairly symmetric.")
```

Exercise 2: Descriptive Statistical Analysis of the Wine Dataset.\
Dataset: A wine quality dataset containing physicochemical properties such as fixed acidity, volatile acidity, citric acid, residual sugar, chlorides, alcohol, sulphates, density, pH, and a sensory quality score (0–10).\
Abstract:

*   What is the typical alcohol content and quality score of the wine samples, and how much variability exists in these attributes across wines categorized as low quality (≤4), medium quality (5–6), and high quality (≥7)?

*   Does the distribution of alcohol content or volatile acidity exhibit skewness or heavy-tailed behaviour, particularly due to the presence of wines with exceptionally high alcohol levels or acidity, indicating concentrated quality-enhancing or quality-degrading effects?

*   Are there any wines that can be classified as statistical outliers based on Z-scores of alcohol content, sulphates, or volatile acidity, and do these extreme values correspond to unusually high or low-quality ratings?

*   Does the presence of wines with very high alcohol content cause a noticeable difference between the mean and median alcohol values, indicating the influence of extreme observations?

*   How large is the spread (standard deviation and interquartile range) of alcohol content, and does it suggest that most wines cluster around a typical alcohol level or are widely dispersed?

```python
df = pd.read_csv('/content/winequality-red.csv')
df.columns = df.columns.str.strip().str.lower()
df.head()
```

```python
print("Columns in dataset:")
print(df.columns)
print("\nDataset info:")
df.info()
```

```python
mean_alcohol = df["alcohol"].mean()
median_alcohol = df["alcohol"].median()
mean_quality = df["quality"].mean()
median_quality = df["quality"].median()
print("Mean Alcohol Content:", mean_alcohol)
print("Median Alcohol Content:", median_alcohol)
print("Mean Quality Score:", mean_quality)
print("Median Quality Score:", median_quality)
```

```python
low_quality = df[df["quality"] <= 4]
medium_quality = df[(df["quality"] >= 5) & (df["quality"] <= 6)]
high_quality = df[df["quality"] >= 7]
print("Low quality wines (≤4):", len(low_quality))
print("Medium quality wines (5–6):", len(medium_quality))
print("High quality wines (≥7):", len(high_quality))
```

```python
print("Alcohol variability by quality category:")
print("Low quality STD:", low_quality["alcohol"].std())
print("Medium quality STD:", medium_quality["alcohol"].std())
print("High quality STD:", high_quality["alcohol"].std())
```

```python
alcohol_skew = stats.skew(df["alcohol"])
alcohol_kurt = stats.kurtosis(df["alcohol"])
volatile_skew = stats.skew(df["volatile acidity"])
volatile_kurt = stats.kurtosis(df["volatile acidity"])
print("Alcohol Skewness:", alcohol_skew)
print("Alcohol Kurtosis:", alcohol_kurt)
print("Volatile Acidity Skewness:", volatile_skew)
print("Volatile Acidity Kurtosis:", volatile_kurt)
```

```python
df["alcohol_zscore"] = stats.zscore(df["alcohol"])
df["sulphates_zscore"] = stats.zscore(df["sulphates"])
df["volatile_zscore"] = stats.zscore(df["volatile acidity"])
df[["alcohol", "sulphates", "volatile acidity", "quality"]].head()
```

```python
outliers = df[
    (np.abs(df["alcohol_zscore"]) > 2) |
    (np.abs(df["sulphates_zscore"]) > 2) |
    (np.abs(df["volatile_zscore"]) > 2)
]
print("Statistical Outliers:")
outliers[["alcohol", "sulphates", "volatile acidity", "quality"]]
```
