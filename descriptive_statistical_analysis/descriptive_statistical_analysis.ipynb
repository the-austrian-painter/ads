```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_excel('/content/Party_Data.xlsx')
```

```python
df['Time'].describe()
```

```python
m=df['Time'].mean()
print(m)
```

```python
med=df['Time'].median()
print(med)
```

```python
mode=df['Time'].mode()
print(mode)
```

```python
harmean=stats.hmean(df['Time'])
print(harmean)
```

```python
m1=df['Time'].max()
m2=df['Time'].min()
```

```python
range=m1-m2
print(range)
```

```python
q3=df['Time'].quantile(0.75)
q1=df['Time'].quantile(0.25)
```

```python
iqr=q3-q1
print(iqr)
```

```python
df.boxplot(['Time'])
```

```python
s=df['Time'].std()
print(s)
```

```python
v=df['Time'].var()
print(v)
```

```python
cv=s/m
print(cv)
```

```python
dq=harmean/m
print(dq)
```

```python
skew=df['Time'].skew()
print(skew)
```

```python
kurtosis=df['Time'].kurtosis()
print(kurtosis)
```

```python
zscore=stats.zscore(df['Time'])
print(zscore)
```

```python
#Conclusion:

# Mean is the average value of the data
# Median is the middle value of the data when arranged in order
# Mode is the value that occurs most frequently in the data
# Harmonic mean is the average used for rates or ratios, giving more weight to smaller values
# Geometric mean is the average that represents the central value of data that grows or changes multiplicatively
# Range is the difference between the highest and lowest values in the data
# Standard deviation shows how much the data values typically deviate from the mean
# Coefficient of variation measures relative variability by comparing the standard deviation to the mean
# Data Quality gives the consistency or reliability of the data values
# Skewness describes whether the data is tilted more to the left or right side
# Kurtosis shows how peaked or flat the data distribution is compared to a normal distribution
```
